# Let's Stemm!

```
# https://www.nltk.org/howto/stem.html

from nltk.stem.snowball import SnowballStemmer
from nltk.stem.porter import PorterStemmer

## https://tartarus.org/martin/PorterStemmer/
stemmer_porter = PorterStemmer()
for word in ['connection', 'connections', 'connective', 'connected', 'connecting']:
    print(stemmer_porter.stem(word))

print('---')

## https://www.geeksforgeeks.org/snowball-stemmer-nlp/
snowball_stemmer_english = SnowballStemmer("english")
for word in ['connection', 'connections', 'connective', 'connected', 'connecting']:
    print(snowball_stemmer_english.stem(word))

# other stemmers
# https://www.geeksforgeeks.org/introduction-to-stemming/
```

```
# differences

for word in ['cared', 'university', 'fairly', 'easily', 'singing', 'sings', 'sung', 'singer', 'sportingly']:
    print(stemmer_porter.stem(word))

print('---')

for word in ['cared', 'university', 'fairly', 'easily', 'singing', 'sings', 'sung', 'singer', 'sportingly']:
    print(snowball_stemmer_english.stem(word))
```

# Bayes theorem

```
import pandas as pd

df_golf = pd.read_csv('golf_data.csv', sep =';')
df_golf
```

```
from sklearn.preprocessing import LabelEncoder

outlook_encoder = LabelEncoder()
outlook = outlook_encoder.fit_transform(df_golf['Outlook'])

temperature_encoder = LabelEncoder()
temperature = temperature_encoder.fit_transform(df_golf['Temperature'])

humidity_encoder = LabelEncoder()
humidity = humidity_encoder.fit_transform(df_golf['Humidity'])

windy_encoder = LabelEncoder()
windy = windy_encoder.fit_transform(df_golf['Windy'])

play_encoder = LabelEncoder()
play = play_encoder.fit_transform(df_golf['Play Golf'])
```

```
print(play_encoder.classes_)
print(play_encoder.transform(play_encoder.classes_))
```

```
print(windy_encoder.classes_)
print(windy_encoder.transform(windy_encoder.classes_))
```

```
print(outlook)
print(temperature)
print(humidity)
print(windy)
print(play)
```

```
df_golf_encoded = pd.DataFrame()
df_golf_encoded['outlook'] = outlook
df_golf_encoded['temperature'] = temperature
df_golf_encoded['humidity'] = humidity
df_golf_encoded['windy'] = windy
df_golf_encoded['play'] = play

df_golf_encoded
```

```
X_train = df_golf_encoded[['outlook', 'temperature', 'humidity', 'windy']].to_numpy()
y_train = np.squeeze(df_golf_encoded[['play']].to_numpy(), axis=1)
```

```
from sklearn.naive_bayes import CategoricalNB

golf_model = CategoricalNB().fit(X_train, y_train)
predicted = golf_model.predict(X_train)

print(np.mean(predicted == y_train))
```

```
X_test = np.array([[0, 0, 0, 0]])
golf_model.predict_proba(X_test)
```

# Four documents and cosine distance

```
from IPython.display import display
```

```
stop_words = open("stopwords_english.txt", "r", encoding="utf-8", newline='').readlines()
stop_words = [word.replace('\n', '') for word in stop_words]
print(stop_words)
```

```
from nltk.tokenize import word_tokenize

d1 = "Natural language processing (NLP) is a subfield of linguistics, computer science, and artificial intelligence concerned with the interactions between computers and human language, in particular how to program computers to process and analyze large amounts of natural language data. The goal is a computer capable of understanding the contents of documents, including the contextual nuances of the language within them."
d2 = "Natural-language understanding (NLU) or natural-language interpretation (NLI) is a subtopic of natural-language processing in artificial intelligence that deals with machine reading comprehension. Natural-language understanding is considered an AI-hard problem. There is considerable commercial interest in the field because of its application to automated reasoning, machine translation, question answering, news-gathering, text categorization, voice-activation, archiving, and large-scale content analysis."
d3 = "Natural language generation (NLG) is a software process that produces natural language output. While it is widely agreed that the output of any NLG process is text, there is some disagreement on whether the inputs of an NLG system need to be non-linguistic. Common applications of NLG methods include the production of various reports, for example weather and patient reports image captions and chatbots."
d4 = "The Battle of Heraklion was fought during World War II on the island of Crete between 20 and 29 May 1941. British, Australian and Greek forces defended the port and the airfield of Heraklion against a German paratrooper attack. The German assault was launched without direct air support and drops occurred over several hours rather than simultaneously."

docs = []
for doc in [d1, d2, d3, d4]:
    words = word_tokenize(doc)
    words = [word.lower() for word in words if word.isalpha()]
    words = [word for word in words if word not in stop_words]
    words = [snowball_stemmer_english.stem(word) for word in words]
    mod_doc = " ".join(words);
    
    print(mod_doc)
    docs.append(mod_doc)
```

```
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer

count_vectorizer = CountVectorizer()
count_matrix = count_vectorizer.fit_transform(docs)
count_dataframe = pd.DataFrame(count_matrix.toarray(), columns = count_vectorizer.get_feature_names_out())
count_dataframe
```

```
d1 = count_dataframe.iloc[0].values
d2 = count_dataframe.iloc[1].values
d3 = count_dataframe.iloc[2].values
d4 = count_dataframe.iloc[3].values
```

```
import numpy as np
from numpy import dot
from numpy.linalg import norm

def cosine(v1, v2):
  val = dot(v1, v2) / ( norm(v1) * norm(v2) )
  return val if np.isnan(val) == False else 0
```

```
print(cosine(d1, d2))
print(cosine(d1, d3))
print(cosine(d1, d4))
```

# Sentiment Analysis

```
import pandas as pd 

df_sentiment = pd.read_csv('sentiment_data.csv', sep ='\t')
```

```
df_sentiment[df_sentiment['sentiment'] == 1]
```

```
df_sentiment[df_sentiment['sentiment'] == 0]
```

```
df_sentiment.to_csv('out.csv', index=False)
```

```
sentiment_normalized = []
for index, row in df_sentiment.iterrows():
    words = word_tokenize(row['text'])
    words = [word.lower() for word in words if word.isalpha()]
    words = [word for word in words if word not in stop_words]
    words = [snowball_stemmer_english.stem(word) for word in words]
    words = [word for word in words if word != None]
    sentiment_normalized.append(" ".join(words))
```

```
df_sentiment['normalized'] = sentiment_normalized
```

```
df_sentiment
```

```
from sklearn.feature_extraction.text import CountVectorizer

count_vectorizer = CountVectorizer(max_features = 1000)
count_train_matrix = count_vectorizer.fit_transform(df_sentiment['normalized'])

count_train_dataframe = pd.DataFrame(count_train_matrix.toarray(), columns = count_vectorizer.get_feature_names_out())
count_train_dataframe
```

```
count_train_result = df_sentiment['sentiment'].to_numpy()
```

```
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(count_train_dataframe, count_train_result, test_size=0.3)
```

```
from sklearn.naive_bayes import MultinomialNB
sentiment_model = MultinomialNB().fit(X_train, y_train)
```

```
import numpy as np

predicted = sentiment_model.predict(X_test)
print(np.mean(predicted == y_test))
```
